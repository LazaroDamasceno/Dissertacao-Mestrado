```python
import geopandas as gpd
from matplotlib import pyplot as plt, gridspec
```

```python
df = gpd.read_file('BR_UF_2024')
```

```python
df.loc[df['NM_REGIA'].isin(['Sudeste']), 'SC1'] = 10
df.loc[df['NM_REGIA'].isin(['Nordeste']), 'SC1'] = 6
df.loc[df['NM_REGIA'].isin(['Sul']), 'SC1'] = 9
df.loc[df['NM_REGIA'].isin(['Norte']), 'SC1'] = 6
df.loc[df['NM_REGIA'].isin(['Centro-oeste']), 'SC1'] = 7
```

```python
df.loc[df['NM_REGIA'].isin(['Sudeste']), 'SC2'] = 11
df.loc[df['NM_REGIA'].isin(['Nordeste']), 'SC2'] = 8
df.loc[df['NM_REGIA'].isin(['Sul']), 'SC2'] = 16
df.loc[df['NM_REGIA'].isin(['Norte']), 'SC2'] = 14
df.loc[df['NM_REGIA'].isin(['Centro-oeste']), 'SC2'] = 15
```

```python
df.loc[df['NM_REGIA'].isin(['Sudeste']), 'SC3'] = 16
df.loc[df['NM_REGIA'].isin(['Nordeste']), 'SC3'] = 10
df.loc[df['NM_REGIA'].isin(['Sul']), 'SC3'] = 7
df.loc[df['NM_REGIA'].isin(['Norte']), 'SC3'] = 11
df.loc[df['NM_REGIA'].isin(['Centro-oeste']), 'SC3'] = 6
```

```python
fig = plt.figure(figsize=(10, 10))
gs = gridspec.GridSpec(nrows=2, ncols=2, figure=fig)

sc1_ax = fig.add_subplot(gs[0, 0])
sc1_plot = df.plot(
    column='SC1',
    legend=True,
    ax=sc1_ax,
    legend_kwds={
        'orientation': 'horizontal'
    }
)
sc1_plot.axis('off')
sc1_plot.set_title('SC1')

sc2_ax = fig.add_subplot(gs[0, 1])
sc2_plot = df.plot(
    column='SC2',
    legend=True,
    legend_kwds={
        'orientation': 'horizontal'
    },
    ax=sc2_ax
)
sc2_plot.axis('off')
sc2_plot.set_title('SC2')

sc3_ax = fig.add_subplot(gs[1, :])
sc3_plot = df.plot(
    column='SC3',
    ax=sc3_ax,
    legend=True
)
sc3_plot.axis('off')
sc3_plot.set_title('SC3')

plt.tight_layout()
plt.savefig('C:/Users/lazar/Master-Dissertation/figuras/mapa_coropletico_tic_domicilios_2024_g2a_2.png', dpi=300, bbox_inches='tight')
plt.show()
```
