```python
import geopandas as gpd
from matplotlib import pyplot as plt, gridspec
```

```python
df = gpd.read_file('BR_UF_2024\\')
```

```python
df.loc[df['NM_REGIA'] == 'Sudeste', 'C1'] = 39
df.loc[df['NM_REGIA'] == 'Nordeste', 'C1'] = 28
df.loc[df['NM_REGIA'] == 'Sul', 'C1'] = 40
df.loc[df['NM_REGIA'] == 'Norte', 'C1'] = 35
df.loc[df['NM_REGIA'] == 'Centro-oeste', 'C1'] = 34
```

```python
df.loc[df['NM_REGIA'] == 'Sudeste', 'C2'] = 39
df.loc[df['NM_REGIA'] == 'Nordeste', 'C2'] = 24
df.loc[df['NM_REGIA'] == 'Sul', 'C2'] = 41
df.loc[df['NM_REGIA'] == 'Norte', 'C2'] = 35
df.loc[df['NM_REGIA'] == 'Centro-oeste', 'C2'] = 30
```

```python
df.loc[df['NM_REGIA'] == 'Sudeste', 'C3'] = 50
df.loc[df['NM_REGIA'] == 'Nordeste', 'C3'] = 63
df.loc[df['NM_REGIA'] == 'Sul', 'C3'] = 46
df.loc[df['NM_REGIA'] == 'Norte', 'C3'] = 52
df.loc[df['NM_REGIA'] == 'Centro-oeste', 'C3'] = 56
```

```python
fig = plt.figure(figsize=(10, 10))
gs = gridspec.GridSpec(nrows=2, ncols=2, figure=fig)

c1_ax = fig.add_subplot(gs[0, 0])
c1_plot = df.plot(
    column='C1',
    legend=True,
    ax=c1_ax
)
c1_plot.set_title('C1')
c1_plot.axis('off')

c2_ax = fig.add_subplot(gs[0, 1])
c2_plot = df.plot(
    column='C2',
    legend=True,
    ax=c2_ax
)
c2_plot.set_title('C2')
c2_plot.axis('off')

c3_ax = fig.add_subplot(gs[1, :])
c3_plot = df.plot(
    column='C3',
    legend=True,
    ax=c3_ax
)
c3_plot.set_title('C3')
c3_plot.axis('off')

plt.tight_layout()
plt.savefig('C:/Users/lazar/Dissertacao-Mestrado-PoderJud-EGDI/figuras/mapa_coropletico_tic_domicilios_2024_g3.png', dpi=300, bbox_inches='tight')
plt.show()
```
