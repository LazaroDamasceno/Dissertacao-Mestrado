```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df1 = gpd.read_file('BR_UF_2024')
```

```python
df1.loc[df1['NM_REGIA'].isin(['Sudeste']), 'Valor'] = 36
df1.loc[df1['NM_REGIA'].isin(['Nordeste']), 'Valor'] = 23
df1.loc[df1['NM_REGIA'].isin(['Sul']), 'Valor'] = 33
df1.loc[df1['NM_REGIA'].isin(['Norte']), 'Valor'] = 30
df1.loc[df1['NM_REGIA'].isin(['Centro-oeste']), 'Valor'] = 23
```

```python
df2 = gpd.read_file('BR_UF_2024')
```

```python
df2.loc[df2['NM_REGIA'].isin(['Sudeste']), 'Valor'] = 37
df2.loc[df2['NM_REGIA'].isin(['Nordeste']), 'Valor'] = 24
df2.loc[df2['NM_REGIA'].isin(['Sul']), 'Valor'] = 32
df2.loc[df2['NM_REGIA'].isin(['Norte']), 'Valor'] = 32
df2.loc[df2['NM_REGIA'].isin(['Centro-oeste']), 'Valor'] = 27
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 8))

ax1 = df1.plot(
    column='Valor',
    legend=True,
    legend_kwds={
        'orientation': 'horizontal'
    },
    ax=ax[0]
)

ax2 = df2.plot(
    column='Valor',
    legend=True,
    legend_kwds={
        'orientation': 'horizontal'
    },
    ax=ax[1]
)

ax[0].set_title('G2-1')
ax[1].set_title('G2-2')
ax[0].axis('off')
ax[1].axis('off')
plt.tight_layout()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/mapa_coropletico_tic_domicilios_2024_g2_1_2.png', dpi=300, bbox_inches='tight')
plt.show()
```
