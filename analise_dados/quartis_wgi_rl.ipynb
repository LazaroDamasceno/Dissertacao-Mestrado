```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
wgi = pd.read_excel('wgidataset.xlsx')
```

```python
rl = wgi.loc[wgi['indicator'] == 'rl'].reset_index(drop=True)
```

```python
rl = rl.replace({
    '..': np.nan
})
```

```python
rl = rl.loc[rl['year'] == 2023].reset_index(drop=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.boxplot(rl, x='pctrank', boxprops=dict(facecolor='white', edgecolor='black'))

ax.set_xlabel('')
ax.spines[['left', 'top', 'right']].set_visible(False)
ax.yaxis.set_visible(False)
plt.tight_layout()
plt.savefig('C:/Users/lazar/Master-Dissertation/figuras/quartis_wgi_rl.png', dpi=300, bbox_inches='tight')
plt.show()
```
