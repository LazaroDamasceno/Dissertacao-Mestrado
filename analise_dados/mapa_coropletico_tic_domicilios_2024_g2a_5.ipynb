```python
import geopandas as gpd
from matplotlib import pyplot as plt, gridspec
```

```python
df = gpd.read_file('BR_UF_2024\\')
```

```python
df.loc[df['NM_REGIA'] == 'Sudeste', 'SC1'] = 19
df.loc[df['NM_REGIA'] == 'Nordeste', 'SC1'] = 7
df.loc[df['NM_REGIA'] == 'Sul', 'SC1'] = 26
df.loc[df['NM_REGIA'] == 'Norte', 'SC1'] = 8
df.loc[df['NM_REGIA'] == 'Centro-oeste', 'SC1'] = 15
```

```python
df.loc[df['NM_REGIA'] == 'Sudeste', 'SC2'] = 3
df.loc[df['NM_REGIA'] == 'Nordeste', 'SC2'] = 2
df.loc[df['NM_REGIA'] == 'Sul', 'SC2'] = 5
df.loc[df['NM_REGIA'] == 'Norte', 'SC2'] = 4
df.loc[df['NM_REGIA'] == 'Centro-oeste', 'SC2'] = 4
```

```python
df.loc[df['NM_REGIA'] == 'Sudeste', 'SC3'] = 12
df.loc[df['NM_REGIA'] == 'Nordeste', 'SC3'] = 10
df.loc[df['NM_REGIA'] == 'Sul', 'SC3'] = 8
df.loc[df['NM_REGIA'] == 'Norte', 'SC3'] = 11
df.loc[df['NM_REGIA'] == 'Centro-oeste', 'SC3'] = 4
```

```python
fig = plt.figure(figsize=(10, 10))
gs = gridspec.GridSpec(nrows=2, ncols=2, figure=fig)

sc1_ax = fig.add_subplot(gs[0, 0])
sc1_plot = df.plot(
    column='SC1',
    legend=True,
    ax=sc1_ax
)
sc1_plot.set_title('SC1')
sc1_plot.axis('off')

sc2_ax = fig.add_subplot(gs[0, 1])
sc2_plot = df.plot(
    column='SC2',
    legend=True,
    ax=sc2_ax
)
sc2_plot.set_title('SC2')
sc2_plot.axis('off')

sc3_ax = fig.add_subplot(gs[1, :])
sc3_plot = df.plot(
    column='SC3',
    legend=True,
    ax=sc3_ax
)
sc3_plot.set_title('SC3')
sc3_plot.axis('off')

plt.tight_layout()
plt.savefig('C:/Users/lazar/Dissertacao-Mestrado-PoderJud-EGDI/figuras/mapa_coropletico_tic_domicilios_2024_g2a_5.png', dpi=300, bbox_inches='tight')
plt.show()
```
