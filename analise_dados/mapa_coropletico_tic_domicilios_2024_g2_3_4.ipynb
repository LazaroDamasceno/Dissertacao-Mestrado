```python
import geopandas as gpd
import  matplotlib.pyplot as plt
```

```python
df1 = gpd.read_file('BR_UF_2024')
```

```python
df1.loc[df1['NM_REGIA'] == 'Sudeste', 'Valor'] = 25
df1.loc[df1['NM_REGIA'] == 'Nordeste', 'Valor'] = 17
df1.loc[df1['NM_REGIA'] == 'Sul', 'Valor'] = 22
df1.loc[df1['NM_REGIA'] == 'Norte', 'Valor'] = 26
df1.loc[df1['NM_REGIA'] == 'Centro-oeste', 'Valor'] = 13
```

```python
df2 = gpd.read_file('BR_UF_2024')
```

```python
df2.loc[df2['NM_REGIA'] == 'Sudeste', 'Valor'] = 28
df2.loc[df2['NM_REGIA'] == 'Nordeste', 'Valor'] = 20
df2.loc[df2['NM_REGIA'] == 'Sul', 'Valor'] = 32
df2.loc[df2['NM_REGIA'] == 'Norte', 'Valor'] = 23
df2.loc[df2['NM_REGIA'] == 'Centro-oeste', 'Valor'] = 18
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

ax1 = df1.plot(
    column='Valor',
    legend=True,
    ax=ax[0]
) 

ax2 = df2.plot(
    column='Valor',
    legend=True,
    ax=ax[1]
) 

ax[0].set_title('G2-3')
ax[1].set_title('G2-4')
ax1.axis('off')
ax2.axis('off')
plt.tight_layout()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/mapa_coropletico_tic_domicilios_2024_g2_3_4.png', dpi=300, bbox_inches='tight')
plt.show()
```
