```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dgi_2020 = pd.DataFrame({
    'Country': ['OCDE', 'Brasil'],
    'Digital by Design': [0.55, 0.54],
    'Data-driven public sector': [0.44, 0.47],
    'Government as a Platform': [0.54, 0.48],
    'Open by Default': [0.64, 0.61],
    'User-Driven': [0.47, 0.54],
    'Proactiveness': [0.42, 0.42],
    'Composite Score': [0.501, 0.519],
}, np.arange(2))
```

```python
dgi_2020 = dgi_2020.melt(
    id_vars='Country',
    value_vars=[
        'Digital by Design',
        'Data-driven public sector',
        'Government as a Platform',	
        'Open by Default',	
        'User-Driven',	
        'Proactiveness',
        'Composite Score'
    ],
    var_name='Dimensoes', 
    value_name='Valores'
)
```

```python
dgi_2023 = pd.DataFrame({
    'Country': ['OCDE', 'Brasil'],
    'Digital by Design': [0.684, 0.622],
    'Data-driven public sector': [0.633, 0.628],
    'Government as a Platform': [0.615, 0.645],
    'Open by Default': [0.525, 0.522],
    'User-Driven': [0.607, 0.639],
    'Proactiveness': [0.567, 0.658],
    'Composite Score': [0.605, 0.619],
}, np.arange(2))
```

```python
dgi_2023 = dgi_2023.melt(
    id_vars='Country',
    value_vars=[
        'Digital by Design',
        'Data-driven public sector',
        'Government as a Platform',	
        'Open by Default',	
        'User-Driven',	
        'Proactiveness',
        'Composite Score'
    ],
    var_name='Dimensoes', 
    value_name='Valores'
)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6), sharey=True)

bar1 = sns.barplot(dgi_2020, y='Dimensoes', x='Valores', hue='Country', palette='Dark2', ax=ax[0])
bar2 = sns.barplot(dgi_2023, y='Dimensoes', x='Valores', hue='Country', palette='Dark2', ax=ax[1])

bar1.legend(title='Organização', loc='upper center', bbox_to_anchor=(0.875, 1.2), ncols=2)
bar1.set_title('2020', fontweight='bold')
bar2.legend().set_visible(False) 
bar2.set_title('2023', fontweight='bold')

for container in zip(bar1.containers, bar2.containers):
    bar1.bar_label(container[0], fmt=lambda x: f'{x:.3f}'.replace('.', ','), padding=3)
    bar2.bar_label(container[1], fmt=lambda x: f'{x:.3f}'.replace('.', ','), padding=3)

for i in range(2):
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].xaxis.set_visible(False)
    ax[i].spines[['top', 'right', 'bottom']].set_visible(False)
    ax[i].xaxis.set_visible(False)

plt.tight_layout()
plt.savefig('C:/Users/lazar/Dissertacao-Mestrado-PoderJud-EGDI/figuras/comparacao_dgi_brasil_mundo_2020_2023.png', dpi=300, bbox_inches='tight')
plt.show()
```
