```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df1 = gpd.read_file('BR_UF_2024')
```

```python
df1.loc[df1['NM_REGIA'].isin(['Sudeste']), 'Valor'] = 33
df1.loc[df1['NM_REGIA'].isin(['Nordeste']), 'Valor'] = 19
df1.loc[df1['NM_REGIA'].isin(['Sul']), 'Valor'] = 38
df1.loc[df1['NM_REGIA'].isin(['Norte']), 'Valor'] = 23
df1.loc[df1['NM_REGIA'].isin(['Centro-oeste']), 'Valor'] = 22
```

```python
df2 = gpd.read_file('BR_UF_2024')
```

```python
df2.loc[df2['NM_REGIA'].isin(['Sudeste']), 'Valor'] = 10
df2.loc[df2['NM_REGIA'].isin(['Nordeste']), 'Valor'] = 6
df2.loc[df2['NM_REGIA'].isin(['Sul']), 'Valor'] = 13
df2.loc[df2['NM_REGIA'].isin(['Norte']), 'Valor'] = 12
df2.loc[df2['NM_REGIA'].isin(['Centro-oeste']), 'Valor'] = 8
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

ax1 = df1.plot(
    column='Valor',
    legend=True,
    legend_kwds={
        'orientation': 'horizontal'
    },
    ax=ax[0]
)

ax2 = df2.plot(
    column='Valor',
    legend=True,
    legend_kwds={
        'orientation': 'horizontal'
    },
    ax=ax[1]
)

ax1.axis('off')
ax2.axis('off')
ax[0].set_title('G2-5')
ax[1].set_title('G2-6')
plt.tight_layout()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/mapa_coropletico_tic_domicilios_2024_g2_5_6.png', dpi=300, bbox_inches='tight')
plt.show()
```
