```python
import geopandas as gpd
```

```python
geo_dados = gpd.read_file('BR_UF_2024')
```

```python
geo_dados.info()
```

```python
geo_dados.loc[geo_dados['NM_REGIA'].isin(['Sudeste']), 'Valor'] = 65
geo_dados.loc[geo_dados['NM_REGIA'].isin(['Nordeste']), 'Valor'] = 49
geo_dados.loc[geo_dados['NM_REGIA'].isin(['Sul']), 'Valor'] = 72
geo_dados.loc[geo_dados['NM_REGIA'].isin(['Norte']), 'Valor'] = 60
geo_dados.loc[geo_dados['NM_REGIA'].isin(['Centro-oeste']), 'Valor'] = 56
```

```python
geo_dados
```

```python
ax = geo_dados.plot(
    column='Valor',
    legend=True,
    figsize=(10, 6)
)

ax.axis('off')

fig = ax.get_figure()
fig.tight_layout()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/mapa_coropletico_tic_domicilios_2024_g1.png', dpi=300, bbox_inches='tight')
fig.show()
```
