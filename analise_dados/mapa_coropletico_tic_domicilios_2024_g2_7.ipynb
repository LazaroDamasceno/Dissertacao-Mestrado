```python
import geopandas as gpd
```

```python
df = gpd.read_file('BR_UF_2024')
```

```python
df.loc[df['NM_REGIA'].isin(['Sudeste']), 'Valor'] = 20
df.loc[df['NM_REGIA'].isin(['Nordeste']), 'Valor'] = 6
df.loc[df['NM_REGIA'].isin(['Sul']), 'Valor'] = 16
df.loc[df['NM_REGIA'].isin(['Norte']), 'Valor'] = 10
df.loc[df['NM_REGIA'].isin(['Centro-oeste']), 'Valor'] = 8
```

```python
ax = df.plot(
    column='Valor',
    legend=True,
    figsize=(10, 6),
    cmap='Dark2'
)

ax.axis('off')

fig = ax.get_figure()

fig.tight_layout()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/mapa_coropletico_tic_domicilios_2024_g2_7.png', dpi=300, bbox_inches='tight')
fig.show()
```
