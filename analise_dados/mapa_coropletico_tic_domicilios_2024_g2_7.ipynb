```python
import geopandas as gpd
```

```python
df = gpd.read_file('BR_UF_2024')
```

```python
df.loc[df['NM_REGIA'] == 'Sudeste', 'Valor'] = 20
df.loc[df['NM_REGIA'] == 'Nordeste', 'Valor'] = 6
df.loc[df['NM_REGIA'] == 'Sul', 'Valor'] = 16
df.loc[df['NM_REGIA'] == 'Norte', 'Valor'] = 10
df.loc[df['NM_REGIA'] == 'Centro-oeste', 'Valor'] = 8
```

```python
ax = df.plot(
    column='Valor',
    legend=True,
    figsize=(10, 6)
)

ax.axis('off')

fig = ax.get_figure()

fig.tight_layout()
fig.savefig('C:/Users/lazar/Dissertacao-Mestrado-PoderJud-EGDI/figuras/mapa_coropletico_tic_domicilios_2024_g2_7.png', dpi=300, bbox_inches='tight')
fig.show()
```
