```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('individuals-using-the-internet.csv')
```

```python
df.info()
```

```python
df = df.drop(columns=[
    'seriesID',
    'seriesCode',
    'seriesName',
    'seriesParent',
    'seriesUnits',
    'entityID',
    'entityIso',
    'dataNote',
    'dataSource',
    'seriesDescription'
])
```

```python
df.info()
```

```python
df = df.rename(columns={
    'entityName': 'Country',
    'dataValue': 'Percentage Use',
    'dataYear': 'Year'
})
```

```python
df.info()
```

```python
df_filtered = df.loc[df['Country'] == "Brazil"]
```

```python
df_filtered.isna().sum()
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(data=df_filtered, x='Year', y='Percentage Use')

sns.despine()
#plt.suptitle('Uso de Internet no Brasil (2000-2024)', fontweight='bold')
plt.ylabel('Percentagem (%)')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```
