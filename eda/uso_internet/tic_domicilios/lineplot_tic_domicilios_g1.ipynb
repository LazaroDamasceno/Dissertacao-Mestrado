```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('tic_domicilios_g1.xlsx')
```

```python
df['Sim'] = df['Sim'].apply(lambda x: x/100.0)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.lineplot(data=df, x='Ano', y='Sim', hue='Regio', palette='tab10')

plt.legend(title='Anos', loc='upper left', bbox_to_anchor=(1.05, 0.5))
sns.despine()
ax.set_ylim(0, 1)
ax.set_ylabel('Escala')
plt.tight_layout()
plt.show()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/internet/crescimento_buscar_egov.png')
```
