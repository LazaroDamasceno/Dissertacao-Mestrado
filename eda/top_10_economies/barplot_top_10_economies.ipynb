```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_imf = pd.read_csv('imf/imf_top_10_economies.csv')
df_worldbank = pd.read_csv('world_bank/worldbank_top_10_economies.csv')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6), sharey=True)

sns.barplot(data=df_imf, y='Country', x='GDP', ax=ax[0], orient='h')
ax[0].set_title('IMF (2025)', fontweight='bold')
ax[0].set_xlabel('GDP in Billion USD')

sns.barplot(data=df_worldbank, y='Country', x='GDP', ax=ax[1], orient='h')
ax[1].set_title('World Bank (2024)', fontweight='bold')
ax[1].set_xlabel('GDP in 10 Trillion USD')

plt.suptitle("GDP of the Top 10 Economies", fontweight='bold')
plt.tight_layout()
plt.show()
```
