```python
import geopandas
import matplotlib.pyplot as plt
```

```python
gdf = geopandas.read_file('brazil_geo.json').sort_values(by='id', ascending=True).reset_index(drop=True)
```

```python
gdf['Regiao'] = None

ne = ['MA', 'PI', 'CE', 'RN', 'PB', 'PE',  'BA', 'SE', 'AL']
co = ['MT', 'MS', 'DF', 'GO']
sul = ['RS', 'SC', 'PR']
sudeste = ['SP', 'MG', 'ES', 'RJ']
norte = ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO']

gdf.loc[gdf['id'].isin(ne), 'Regiao'] = 'Nordeste'
gdf.loc[gdf['id'].isin(co), 'Regiao'] = 'Centro-Oeste'
gdf.loc[gdf['id'].isin(sul), 'Regiao'] = 'Sul'
gdf.loc[gdf['id'].isin(sudeste), 'Regiao'] = 'Sudeste'
gdf.loc[gdf['id'].isin(norte), 'Regiao'] = 'Norte'                    
```

```python
gdf['uso_egov'] = None
 
gdf.loc[gdf['id'].isin(ne), 'uso_egov'] = 49
gdf.loc[gdf['id'].isin(co), 'uso_egov'] = 56
gdf.loc[gdf['id'].isin(sul), 'uso_egov'] = 72
gdf.loc[gdf['id'].isin(sudeste), 'uso_egov'] = 65
gdf.loc[gdf['id'].isin(norte), 'uso_egov'] = 60
```

```python
gdf.dtypes
```

```python
gdf['uso_egov'] = gdf['uso_egov'].astype(float)
```

```python
gdf.plot(column='uso_egov',
        cmap='Dark2',
        legend=True,
        figsize=(10, 6),
        vmin=0,
        vmax=100
)

plt.tight_layout()
plt.savefig('C:/Users/lazar/Master-Dissertation/figuras/mapa_coropleto_tic_domicilio_g1.png', dpi=300, bbox_inches='tight')
plt.show()
```
