```python
edi = pd.read_csv('electoral-democracy-index.csv')
```

```python
edi = edi.iloc[:, [0,2,3]]
```

```python
edi = edi[edi['Entity'].isin(countries)].reset_index(drop=True)
```

```python


edi = edi[edi['Year'].isin(itertools.chain(range(2003, 2006), range(2008, 2025, 2)))].reset_index(drop=True)
```

```python
df = egdi.copy()
```

```python
df['Índice de Democrácia Eleitoral'] = edi['Electoral democracy index (central estimate)']
```

```python
corr = df.corr(method='pearson')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.heatmap(data=corr, annot=True)

plt.tight_layout()
plt.show()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/egdi/correlacao3.png')
```
