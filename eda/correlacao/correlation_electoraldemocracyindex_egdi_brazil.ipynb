```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
```

```python
country = ["Brazil"]
```

```python
egdi_2003 = pd.read_csv('EGOV_DATA_2003.csv')
egdi_2004 = pd.read_csv('EGOV_DATA_2004.csv')
egdi_2005 = pd.read_csv('EGOV_DATA_2005.csv')
egdi_2008 = pd.read_csv('EGOV_DATA_2008.csv')
egdi_2010 = pd.read_csv('EGOV_DATA_2010.csv')
egdi_2012 = pd.read_csv('EGOV_DATA_2012.csv')
egdi_2014 = pd.read_csv('EGOV_DATA_2014.csv')
egdi_2016 = pd.read_csv('EGOV_DATA_2016.csv')
egdi_2018 = pd.read_csv('EGOV_DATA_2018.csv')
egdi_2020 = pd.read_csv('EGOV_DATA_2020.csv')
egdi_2022 = pd.read_csv('EGOV_DATA_2022.csv')
egdi_2024 = pd.read_csv('EGOV_DATA_2024.csv')
```

```python
egdi = pd.concat([
    egdi_2003,
    egdi_2004,
    egdi_2005,
    egdi_2008,
    egdi_2010,
    egdi_2012,
    egdi_2014,
    egdi_2016,
    egdi_2018,
    egdi_2020,
    egdi_2022,
    egdi_2024
]).reset_index(drop=True)
```

```python
egdi = egdi[egdi['Country Name'].isin(country)].reset_index(drop=True)
egdi = egdi.iloc[:, range(3, 8)]
```

```python
edi = pd.read_csv('electoral-democracy-index.csv')
```

```python
edi = edi.iloc[:, [0,2,3]]
```

```python
edi = edi[edi['Entity'].isin(country)].reset_index(drop=True)
```

```python


edi = edi[edi['Year'].isin(itertools.chain(range(2003, 2006), range(2008, 2025, 2)))].reset_index(drop=True)
```

```python
df = egdi.copy()
```

```python
df['Índice de Democrácia Eleitoral'] = edi['Electoral democracy index (central estimate)']
```

```python
corr = df.corr(method='pearson')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.heatmap(data=corr, annot=True)

plt.tight_layout()
plt.show()
fig.savefig('C:/Users/lazar/Master-Dissertation/figuras/egdi/correlacao8.png')
```
